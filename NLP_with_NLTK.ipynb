```python
pdf = open('Introduction+to+natural+language+processing.pdf','rb')
reader = PyPDF2.PdfReader(pdf)
print(len(reader.pages))

```

```python
from PyPDF2 import PdfReader

first_page = reader.pages[4]
page_text = first_page.extract_text().replace('\n',' ')


page = reader.pages[4]
parts = []


def visitor_body(text, cm, tm, fontDict, fontSize):
    y = tm[5]
    if y > 50 and y < 720:
        parts.append(text)


page.extract_text(visitor_text=visitor_body)
text_body = "".join(parts)

text_body
```

### **No.1 Tokenization**

```python
from nltk import sent_tokenize
from nltk import word_tokenize
import nltk
# nltk.download('punkt') --if you encountered error just activate this or just leave it

tokenize_sent = sent_tokenize(str(text_body))
tokenize_sent
```

```python
tokenize_word = word_tokenize(str(text_body))
tokenize_word
```

### **No.2 Lemmatzation**

```python
# first letes downlaode the wordnew
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmati = WordNetLemmatizer()
def the_lemmatizer(data):
    for i in data:
        print(i,':', lemmati.lemmatize(i.replace('•',' ')))
the_lemmatizer(tokenize_word)
```

### **No.3 Stop words**

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words())
```

```python
# let's print the words
for word in tokenize_word:
    if word in stop_words:
        print(word)
```
